### Логистическая регрессия | 2022

Обучена логистическая регрессия с масштабированием, оценивается важность признаков.

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_wine
data = load_wine()
```

```python
dataset = data['data'][data['target']<2]
target = data['target'][data['target']<2]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data_t = scaler.fit_transform(dataset)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(data_t, target)
```

```python
sorted_weights = sorted(zip(clf.coef_.ravel(), data['feature_names']), reverse=True)
weights = [x[0] for x in sorted_weights]
features = [x[1] for x in sorted_weights]
df = pd.DataFrame({'features': features, 'weights':weights})
ax = df.plot.barh(x='features', y='weights', rot=0, )
```

Обучена логистическая регрессия без масштабирования, оценивается важность признаков.

Обучается и оценивается логистическая регрессия на первом варианте векторизации.

```python
spam = pd.read_csv('SMSSpamCollection.csv', sep='\t', names=['target', 'data'])
X_train, y_train = spam['data'][:4000], spam['target'][:4000]
X_test, y_test = spam['data'][4000:], spam['target'][4000:]

vectorizer_tf_idf = TfidfVectorizer()

X_train_tfidf = vectorizer_tf_idf.fit_transform(X_train)
X_test_tfidf = vectorizer_tf_idf.transform(X_test)
X_test_tfidf.shape

clf4 = LogisticRegression()
clf4.fit(X_train_tfidf, y_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(clf4.predict(X_test_tfidf), y_test)
```
